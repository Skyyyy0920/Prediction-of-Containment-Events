# 以经停病例数作为输入变量

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
path = 'C:/Users/new/Desktop/Prediction-of-Containment-Events/数据/Baidu_Trajectory_Geocoding.csv'
data = pd.read_csv(path)
```

```python
data.drop(['原始文本', '来源', '风险等级', '存疑'], axis=1, inplace=True) # 去掉不需要的列
```

```python
# 从实际来说，患者经停的场所为下方这几类时并不具有传染性，这部分数据应该除去
rows_index = data[(data['场所'] == '居家') | (data['场所'] == '定点医院') | (data['场所'] == '隔离点') | (data['场所'] == '定点酒店')].index.tolist()
len(rows_index)
data.drop(index=rows_index, axis=0, inplace=True)
```

```python
# 保存为csv
output_path='C:/Users/new/Desktop/train_X.csv'
data.to_csv(output_path,sep=',',index=False,header=True,encoding='utf_8_sig')
```

```python
path = 'C:/Users/new/Desktop/Prediction-of-Containment-Events/数据/Baidu_Patient_Geocoding.csv'
data_pat = pd.read_csv(path)
```

```python
data_pat.drop(['来源1','来源2'], axis=1, inplace=True)
```

```python
data_pat.head()
data_pat.info()
```

```python
# 找出所有street(road)
street_set = set()
# 轨迹数据集中涉及到的street
for index,row in data.iterrows():
    if row['road1'] is not np.NaN:
        street_set.add(row['road1'])
    if row['road2'] is not np.NaN:
        street_set.add(row['road2'])
    if row['road3'] is not np.NaN:
        street_set.add(row['road3'])
# 患者数据集中涉及到的street
for index,row in data_pat.iterrows():
    if row['road1'] is not np.NaN:
        street_set.add(row['road1'])
    if row['road2'] is not np.NaN:
        street_set.add(row['road2'])
    if row['road3'] is not np.NaN:
        street_set.add(row['road3'])
```

```python
street_set
```

```python
len(street_set)
```

```python
# 生成时间序列
start_timestamp = pd.Timestamp(2021,12,23)
end_timestamp = pd.Timestamp(2022,2,1)
time_series = pd.date_range(start_timestamp,end_timestamp)
```

```python
time_series
```

```python
train_X = pd.DataFrame(columns=('日期', 'district', 'street', '经停病例数'))
length = len(train_X.index)
# 找到对应的区，感觉有更好的算法，但是目前想不出来
for place in street_set:
    rows_index = data[(place == data['road1']) | (place == data['road2']) | (place == data['road3'])].index.tolist()
    if rows_index:  # 不是在轨迹数据就是在患者数据中找   
        district = data.loc[rows_index[0], 'district']
    else:
        rows_index = data_pat[(place == data_pat['road1']) | (place == data_pat['road2']) | (place == data_pat['road3'])].index.tolist()
        district = data_pat.loc[rows_index[0], 'district']
    for time in time_series:
        train_X.loc[length] = [time, district, place, 0]
        length += 1
```

```python
train_X.head()
train_X.info()
```

```python
for index,row in data.iterrows():
    startdatetime = row['起始日期']
    rows_index = train_X[(train_X['日期'] == startdatetime) & ((train_X['street'] == row['road1']) | (train_X['street'] == row['road2']) | (train_X['street'] == row['road3']))].index.tolist()
    if rows_index:
        for r in rows_index:
            train_X.loc[r, '经停病例数'] += 1
    # else:
        # print("error")
        # train_X.loc[len(train_X.index)] = [datetime.fromtimestamp(startdatetime.timestamp()), place, 1]
```

```python
# 保存为csv
output_path='C:/Users/new/Desktop/train_X.csv'
train_X.to_csv(output_path,sep=',',index=False,header=True,encoding='utf_8_sig')
```
