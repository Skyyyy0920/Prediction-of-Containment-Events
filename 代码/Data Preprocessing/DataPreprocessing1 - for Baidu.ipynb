# 新增患者数量、全市患者数量、各区患者数量作为输入变量

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
path = 'C:/Users/new/Desktop/Prediction-of-Containment-Events/数据/Baidu_Trajectory_Geocoding.csv'
data = pd.read_csv(path)
```

```python
data.drop(['原始文本', '来源', '风险等级', '存疑'], axis=1, inplace=True) # 去掉不需要的列
```

```python
data.head()
data.info()
```

```python
path = 'C:/Users/new/Desktop/Prediction-of-Containment-Events/数据/Baidu_Patient_Geocoding.csv'
data_pat = pd.read_csv(path)
```

```python
data_pat.drop(['来源1','来源2'], axis=1, inplace=True)
```

```python
data_pat.head()
data_pat.info()
```

```python
path = 'C:/Users/new/Desktop/Prediction-of-Containment-Events/数据/street_dataset.csv'
train = pd.read_csv(path)
```

```python
train["患者数量"] = 0
train["全市患者数量"] = 0
train["所在区患者数量"] = 0
```

```python
for index,row in train.iterrows():
    train.loc[index, '日期'] = datetime.strptime(row['日期'], "%Y-%m-%d")
```

```python
train
```

```python
# 找出所有区
district_set = set()
for index,row in data.iterrows():
    if row['district'] is not np.NaN:
        district_set.add(row['district'])
district_set = list(district_set)
```

```python
district_set
```

```python
# 生成时间序列
start_timestamp = pd.Timestamp(2021,12,23)
end_timestamp = pd.Timestamp(2022,2,1)
time_series = pd.date_range(start_timestamp,end_timestamp)
```

```python
time_series
len(time_series)
type(time_series)
```

```python
l = list(time_series)
```

```python
length = 41  # 这里解释一下，这个东西是时间序列长度为41，即统计了41天，因此这里取41，拿下面的index模41，就能得到对应区的对应时间患者数量
total_num = list(np.zeros(length)) # 全市每天的患者数量
district_num = {}  # 每个区每天的患者数量
for i in district_set:
    district_num[i] = list(np.zeros(length))
```

```python
startdatetime = datetime.strptime('2022-01-07', "%Y-%m-%d")
for index,row in data_pat.iterrows():
    pat_id = row['患者序号']
    r1, r2, r3 = row['road1'], row['road2'], row['road3']
    district = row['district']
    # 找出患者检测出阳性的时间
    rows = data[(data['病例号'] == pat_id)]
    d = rows.iloc[len(rows)-1, 1]  # 这里取最后一天，因为患者的轨迹数据是从患者被检测出阳性那一天往前14天的数据，那么轨迹的最后一天就是患者被检测出阳性的那天
    if d is not np.NaN:
        d = datetime.strptime(d, "%Y/%m/%d")
        if startdatetime < d:
            startdatetime = d
    ind = l.index(startdatetime)
    # 统计全市患者人数
    for time_lag in range(7):  # 这里假设一个患者有7天的影响
        if ind + time_lag < length:
            total_num[ind + time_lag] += 1
    # 统计各区患者人数
    if district is not np.NaN:
        for time_lag in range(7):
            if ind + time_lag < length:
                district_num[district][ind + time_lag] += 1
    # 统计每个街道患者人数
    rows_index = train[(train['日期'] == startdatetime) & ((train['street'] == r1) | (train['street'] == r2) | (train['street'] == r3))].index.tolist()
    if rows_index:
        for r in rows_index:
            for time_lag in range(7):
                if r + time_lag < len(train):
                    street = train.loc[r+time_lag, 'street']
                    if street == r1 or street == r2 or street == r3:
                        train.loc[r+time_lag, '患者数量'] += 1
```

```python
total_num
```

```python
district_num
```

```python
i = 0
for index,row in train.iterrows():
    district = row['district']
    train.loc[index, '全市患者数量'] = total_num[i]
    train.loc[index, '所在区患者数量'] = district_num[district][i]
    i += 1
    i = i % length
```
