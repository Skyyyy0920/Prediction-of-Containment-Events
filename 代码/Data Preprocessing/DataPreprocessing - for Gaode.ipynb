# 以经停病例数作为输入变量

```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
path = 'C:/Users/new/Desktop/Prediction-of-Containment-Events/数据/Geocoding.csv'
data = pd.read_csv(path)
```

```python
data.drop(['原始文本', '来源', '风险等级', '存疑'], axis=1, inplace=True) # 去掉不需要的列
```

```python
data.head()
```

```python
# 找出所有区
region_set = set()
for index,row in data.iterrows():
    if row['township'] is not np.NaN:
        region_set.add(row['township'])
```

```python
region_set
```

```python
len(region_set)
```

```python
# 生成时间序列
start_timestamp = pd.Timestamp(2021,12,23)
end_timestamp = pd.Timestamp(2022,2,1)
time_series = pd.date_range(start_timestamp,end_timestamp)
```

```python
time_series
```

```python
train_X = pd.DataFrame(columns=('日期', '所在街道', '经停病例数'))
length = len(train_X.index)
for place in region_set:
    for time in time_series:
        train_X.loc[length] = [time, place, 0]
        length += 1
```

```python
train_X
```

```python
for index,row in data.iterrows():
    startdatetime = row['起始日期']
    township = row['township']
    rows_index = train_X[(train_X['日期'] == startdatetime) & (train_X['所在街道'] == township)].index.tolist()
    if rows_index:
        for r in rows_index:
            train_X.loc[r, '经停病例数'] += 1
            # print(train_X.loc[r])
    # else:
        # print("error")
        # train_X.loc[len(train_X.index)] = [datetime.fromtimestamp(startdatetime.timestamp()), place, 1]
```

```python
output_path='C:/Users/new/Desktop/train_X.csv'
train_X.to_csv(output_path,sep=',',index=False,header=False,encoding='utf_8_sig')
```
