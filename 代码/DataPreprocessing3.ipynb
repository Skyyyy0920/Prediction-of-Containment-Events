```python
path = 'C:/Users/new/Desktop/疫情演化系统/数据/train_township.csv'
train = pd.read_csv(path)
```

```python
# 找出所有区
region_set = set()
for index,row in data.iterrows():
    if row['所在区'] is not np.NaN:
        region_set.add(row['所在区'])
region_set = list(region_set)
```

```python
region_set
```

```python
# 创建一个字典，存储每个township对应的区
dictionary = {}
for index,row in data.iterrows():
    township = row['township']
    region = row['所在区']
    if township is not np.NaN:
        if township not in dictionary.keys():
            dictionary[township] = region
for index,row in data_pat.iterrows():
    township = row['township']
    region = row['所在区']
    if township is not np.NaN:
        if township not in dictionary.keys():
            dictionary[township] = region
```

```python
# 创建另一个字典，用来存储每个区的患者数量
length = 41 # 这里解释一下，这个东西是时间序列长度为41，即统计了41天，因此这里取41，拿下面的index模41，就能得到对应区的对应时间患者数量
dict_num = {}
for i in region_set:
    dict_num[i] = list(np.zeros(length))
```

```python
# 计算每个区在这一段时间每天的患者数量，假设一个患者有14天的影响
for index,row in train.iterrows():
    township = row['township']
    num = row['患者数量']
    region = dictionary[township]
    ind = index % length
    dict_num[region][ind] += num
```

```python
dict_num
```

```python
# 将数据填入原表格
train['各区患者数量'] = 0
i = 0
for index,row in train.iterrows():
    township = row['township']
    train.loc[index, '各区患者数量'] = dict_num[dictionary[township]][i]
    i = i + 1
    i = i % length
    
# startdatetime = datetime.strptime('2022-01-07', "%Y-%m-%d")
# for index,row in data_pat.iterrows():
#     pat_id = row['患者序号']
#     township = row['township']
#     region = row['所在区']
#     # 找出患者检测出阳性的时间
#     rows = data[(data['病例号'] == pat_id)]
#     d = rows.iloc[len(rows)-1, 1]
#     if d is not np.NaN:
#         d = datetime.strptime(d, "%Y/%m/%d")
#         if startdatetime < d:
#             startdatetime = d
#     rows_index = train[(train['日期'] == startdatetime) & (train['镇区'] == township)].index.tolist()
#     township
#     rows_index
#     if rows_index:
#         for r in rows_index:
#             # 这里假设一个患者有14天的影响
#             for time_lag in range(14):
#                 train.loc[r+time_lag, '患者数量'] += 1
```
