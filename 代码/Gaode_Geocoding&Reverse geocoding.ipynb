## 基于高德地图地理编码与逆地理编码API的数据地点层次化处理

```python
API_key = "947c7f60f731b54bd6474ce18d5d894b"  # 使用高德地图地理编码API所需的key,这里我申请的是个人key,每日只能查询5000条数据
city = "天津"  # 默认对天津的地点进行地理编码和逆编码
city = urllib.parse.quote(city)  # 这里要注意,按照标准,URL只允许一部分ASCII字符,其他字符（如汉字）是不符合标准的，此时就要进行编码,下方的address同样需要
```

```python
def subway(s):
    """
    这个函数属于无奈之举，比如地铁5号线南翠屏站，他就是没法给你返回正常结果，在高德里换成南翠屏(地铁站)会好一些
    """
    ditie = "地铁"
    haoxian = "号线"
    gongjiao = "公交站"
    address = ""
    if ditie in s and haoxian in s:
        address = s[s.find(haoxian) + 2:-1] + "(地铁站)"
    elif gongjiao in s:
        address = s[:s.find(gongjiao)] + "(公交站)"
    else:
        address = s
    return address
```

```python
place = []
for index,row in data.iterrows():
    address = subway(row['场所'])
#     address = "天津市" + address
    if address is not np.NaN:
        address = urllib.parse.quote(address)
        geo_url = "https://restapi.amap.com/v3/geocode/geo?key={}&address={}&city={}".format(API_key,address,city)

        req = urllib.request.Request(geo_url)  # 将url对象转化为request对象
        ret = urllib.request.urlopen(req)  # 发送请求，打开url，也可以直接传入url
        result = ret.read().decode('utf-8')

        result = eval(result)
        if result['status'] == '1':  # 有时候API会有一些没法返回正常的查询结果，比如乐齐国际影城，不知道为啥返回不了正常结果，status=0
            location = result['geocodes'][0]['location']
            longitude, latitude = eval(location)  # 地点对应经纬度

            regeo_url = 'https://restapi.amap.com/v3/geocode/regeo?key={}&location={}&extensions=all'.format(API_key,location)  # extensions 参数取值为 all 时会返回基本地址信息、附近 POI 内容、道路信息以及道路交叉口信息
            req = urllib.request.Request(regeo_url)  # 将url对象转化为request对象
            ret = urllib.request.urlopen(req)  # 发送请求，打开url，也可以直接传入url
            result = ret.read().decode('utf-8')

            result = eval(result)
            if result['status'] == '1':
                formatted_address = result['regeocode']['formatted_address']  # 完整地址
                district = result['regeocode']['addressComponent']['district']  # 所在区
                street = result['regeocode']['addressComponent']['streetNumber']['street']  # 所在街道
                street_number = result['regeocode']['addressComponent']['streetNumber']['number']  # 所在街道门牌号
                township = result['regeocode']['addressComponent']['township']  # 所在乡镇一级
#                 business = result['regeocode']['addressComponent']['businessAreas'][0]['name']  # 所在商圈
                place_around = result['regeocode']['pois']  # 附近地点，list类型
                road1, road2, road3 = '', '', ''
                if len(result['regeocode']['roads']) == 3:
                    road1 = result['regeocode']['roads'][0]['name']  # 附近的路
                    road2 = result['regeocode']['roads'][1]['name']
                    road3 = result['regeocode']['roads'][2]['name']
                elif len(result['regeocode']['roads']) == 2:
                    road1 = result['regeocode']['roads'][0]['name']  # 附近的路
                    road2 = result['regeocode']['roads'][1]['name']
                elif len(result['regeocode']['roads']) == 1:
                    road1 = result['regeocode']['roads'][0]['name']  # 附近的路

                if formatted_address:  # 主要是防止有些地址返回的是空集
                    data.loc[index, 'formatted_address'] = formatted_address
                if district:
                    data.loc[index, 'district'] = district
                if street:
                    data.loc[index, 'street'] = street
                if street_number:
                    data.loc[index, 'street_number'] = street_number
                if longitude:
                    data.loc[index,'经度'] = longitude
                if latitude:
                    data.loc[index,'纬度'] = latitude
#                 if business:
#                     data.loc[index, 'business'] = business
                if place_around:
                    data.loc[index, 'place_around'] = index  # 这里只存索引，另外拿一张表存附近地区信息，根据索引查找
                    place.append(place_around)
                if road1:
                    data.loc[index, 'road1'] = road1
                if road2:
                    data.loc[index, 'road2'] = road2
                if road3:
                    data.loc[index, 'road3'] = road3
```
